```python
import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(0)

# Define the number of samples
num_samples = 10000

# Simulate academic performance (GPA) on a 5.0 scale with 2 decimal places
mean_gpa = 3.0
std_dev_gpa = 0.5
gpa = np.random.normal(mean_gpa, std_dev_gpa, num_samples)
gpa = np.clip(gpa, 1.0, 5.0)  # Clip GPA values to be within a 1.0-5.0 range
gpa = np.round(gpa, 2)  # Round GPA values to two decimal places

# Simulate the binary target variable (1 for success, 0 for failure)
success_rate = 0.6  # Adjust this to set the success rate
success = np.random.choice([1, 0], num_samples, p=[success_rate, 1 - success_rate])

# Simulate additional features
education_levels = np.random.choice(['High School', 'Bachelors', 'Masters', 'PhD'], num_samples)
test_score = np.random.randint(0, 101, num_samples)  # Simulate test scores on a 0-100 scale
num_programming_languages = np.random.randint(0, 6, num_samples)  # Number of programming languages known
internship_type = np.random.choice(['Backend', 'Frontend', 'Machine Learning'], num_samples)  # Type of internship
farm_experience = np.random.randint(0, 11, num_samples)  # Years of farm-related experience
research_publications = np.random.randint(0, 6, num_samples)  # Number of research publications

# Change "Geographic_Region" to "Continent"
continent = np.random.choice(['North America', 'South America', 'Europe', 'Asia', 'Africa', 'Australia'], num_samples)

# Create a DataFrame to store the data
data = pd.DataFrame({
    'GPA': gpa,
    'Skills': num_programming_languages,
    'Internship_Success': success,
    'Education_Level': education_levels,
    'Test_Score': test_score,
    'Continent': continent,
    'Internship_Type': internship_type,
    'Farm_Experience': farm_experience,
    'Research_Publications': research_publications
})

# Save the dataset to a CSV file
data.to_csv('climdes_intern_dataset.csv', index=False)
```
